```python
#imports
import os
import pandas as pd
import numpy as np
from math import sin, cos, sqrt, atan2, radians
```

```python
filename = os.path.join(os.getcwd(), "hurdat2_sample.csv")
df = pd.read_csv(filename)
```

```python
df.head(10)
```

```python
#shorten column names and strip extra spaces from Name
df.rename(columns={"Unnamed: 0": "Name", "Storm Name/Record Identifier": "Record Identifier", "Number of Entries/Status": "Status of System"}, inplace=True)
df["Name"] = df["Name"].str.strip()
```

```python
#move name from Time column to Name column
df.insert(1, "Num Entries", df[df["Record Identifier"].astype(str).str.isdigit()]["Record Identifier"])
df["Name"] = df.loc[df["Year"].str.startswith("AL")]["Time"]
df["Name"] = df["Name"].fillna(method='ffill')
#propogate num entries
df["Num Entries"] = df["Num Entries"].fillna(method='ffill')
#delete header rows
df = df.drop(df[df["Year"].str.startswith("AL")].index)
```

```python
nan_count = np.sum(df.isnull(), axis = 0)
nan_count
```

```python
#delete time entries that are not 12:00 or 6:00 am/pm
time_vals = df["Time"].unique()
time_vals = np.delete(time_vals, [0, 1, 2, 3])
df = df.replace(time_vals, np.nan).dropna()
```

```python
np.sum(df.isnull(), axis = 0)
```

```python
#drop rows that do not have Wind Radii information
df = df.replace(-999.0, np.nan)
df = df.drop(df[df["34-knot Wind Radii SW"].isnull()].index)
```

```python
#get rows for specified months/years
months = ['9']
years = ['2020', '2021', '2022', '2023']
newDf = pd.DataFrame()
for y in years:
    for m in months:
        dfAdd = df[df["Year"].str.startswith(m)]
        dfAdd = dfAdd[dfAdd["Year"].str.endswith(y)]
        newDf = pd.concat([newDf, dfAdd], ignore_index=True)

df = newDf
```

Calculate speed of the hurricane/storm:

```python
def calc_distance(lat1, lon1, lat2, lon2):
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    R = 3958.8
    a = (sin(dlat / 2))**2 + cos(lat1) * cos(lat2) * (sin(dlon / 2))**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = R * c
    return distance
```

```python
#create new columns to perform calculations
df['lat1'] = df['LAT'].shift(1)
df['lon1'] = df['LON'].shift(1)
df['lat1'].fillna(value=0, inplace=True)
df['lon1'].fillna(value=0, inplace=True)
df = df.reset_index(drop=True)
lat1Vals = df["lat1"]
lat2Vals = df["LAT"]
lon1Vals = df["lon1"]
lon2Vals = df["LON"]
```

```python
#calculate distance and add it to dataframe
dist = []
len = lat1Vals.size
for index in range(len):
    dist.append(calc_distance(radians(lat1Vals[index]), radians(lon1Vals[index]), radians(lat2Vals[index]), radians(lon2Vals[index])))
dist = pd.Series(dist)
df["Distance"] = dist
df["Distance"] = df["Distance"] / 6
#drop columns used for calculations
df = df.drop(columns=['lat1', 'lon1'], axis=1)
df
```

```python
#make the first entry's speed null
for index in range(len - 1):
    if df.loc[index, 'Name'] != df.loc[index + 1, 'Name']:
        df.loc[index + 1, 'Distance'] = np.nan
df.loc[0, 'Distance'] = np.nan
df = df.rename(columns={"Distance":"Speed mph"})
```

```python
#export cleaned data
df.to_csv('hurdat_sep20_23.csv')
```
