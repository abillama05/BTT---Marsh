```python
# we need to convert the vessel's BaseDateTime column to panda's datetime format so that we can merge datasets
vessel_df['BaseDateTime'] = pd.to_datetime(vessel_df['BaseDateTime'])
```

```python
# TEMPORARILY: dropped rows with NaN values in 'Time' column
# hurricane_df = hurricane_df.dropna(subset=['Time'])
```

```python
# TEMPORARILY: removed rows where 'Time' is not a number or contains invalid data
# hurricane_df = hurricane_df[hurricane_df['Time'].str.strip().str.isdigit()]
```

```python
# this is just ensuring that all the values in the 'Time' column is 4 digits
hurricane_df['Time'] = hurricane_df['Time'].apply(lambda x: f'{int(x):04d}')  # Ensure it's 4 digits
```

```python
hurricane_df
```

```python
# we need to convert the hurricane's BaseDateTime column to panda's datetime format so that we can merge datasets
# we need to combine the 'Year' and 'Time' columns in the hurricane dataset to form a proper datetime
hurricane_df['hurricane_datetime'] = pd.to_datetime(hurricane_df['Year'] + ' ' + hurricane_df['Time'], format='%m/%d/%Y %H%M')
```

```python
hurricane_df
```

```python
# we round the vessel data to the nearest 6 hours so that it can match the hurricane data
# vessel_df['rounded_time'] = vessel_df['BaseDateTime'].dt.round('6H')
```

```python
# this is where we merge the two datasets on the rounded time
merged_df = pd.merge(vessel_df, hurricane_df, left_on='BaseDateTime', right_on='hurricane_datetime', how='inner')
```
